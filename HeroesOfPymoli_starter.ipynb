## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
bins = [0,9,14, 19, 24, 29, 34, 39,100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
age_grouped = purchase_data.groupby("Age Group") 
count_age = age_grouped["Purchase ID"].count() 
price_sum = age_grouped["Price"].sum()
avg_purchase_price = price_sum/count_age
unique_age = age_grouped['SN'].nunique()
avg_per_person = price_sum / unique_age

age_purchase =pd.DataFrame({"Purchase Count":count_age, 
                    "Average Purchase Price": 
                    avg_purchase_price.astype(float).map("${:,.2f}".format),
                    "Total Purchase Value": price_sum.astype(float).map("${:,.2f}".format),
                    "Avg Total Purchase per Person": 
                    avg_per_person.astype(float).map("${:,.2f}".format)})
age_purchase
#3
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns

* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#??why double the [] 
items = purchase_data[["Item ID", "Item Name", "Price"]]
items.head(10)
```

```python
items_grouped = items.groupby(["Item ID","Item Name"])
item_count = items_grouped.size()
price_sum = items_grouped["Price"].sum()
price = price_sum/item_count
```

```python
item_popular1 =pd.DataFrame({"Purchase Count":item_count, 
                    "Item Price": price,
                    "Total Purchase Value": price_sum})
item_popular2 =item_popular1.sort_values("Purchase Count",ascending=False)
item_popular2["Item Price"]=item_popular2["Item Price"].map("${:,.2f}".format) 
item_popular2["Total Purchase Value"]=item_popular2["Total Purchase Value"].map("${:,.2f}".format) 
item_popular2 = item_popular2.iloc[0:5]
item_popular2 
               
```
